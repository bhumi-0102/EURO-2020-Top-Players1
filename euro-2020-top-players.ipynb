# EURO-2020 TOP PLAYERS 

The 2020 UEFA European Football Championship, commonly referred to as UEFA Euro 2020 or simply Euro 2020, was the 16th UEFA European Championship, the quadrennial international men's football championship of Europe organised by the Union of European Football Associations (UEFA). To celebrate the 60th anniversary of the European Championship competition, UEFA president Michel Platini declared that the tournament would be hosted in several nations as a "romantic" one-off event, with 11 cities in 11 UEFA countries each providing venues for the tournament. Defending champions Portugal, who won Euro 2016 in France, were eliminated in the round of 16 by Belgium. Italy won their second European Championship title by beating England on penalties in the final following a 1â€“1 draw after extra time.

Due to the COVID-19 pandemic in Europe during 2020, the tournament was postponed to summer 2021,while retaining the name UEFA Euro 2020 and host venues. Alongside special rules regarding COVID-19, UEFA also allowed more substitutions and implemented video assistant referee (VAR) for the first time.

Initially, 13 venues were chosen for the tournament but two were later dropped. Brussels was dropped in December 2017 after the building of the city's Eurostadium was abandoned, while Dublin was dropped in April 2021 because there was no guarantee that spectators could attend. Spain originally intended to use Bilbao as a host venue but later changed it to Seville to allow for spectators at matches. UEFA chose Stadio Olimpico in Rome to host the opening match between Italy and Turkey, while Wembley Stadium in London was selected as a semi-final and final venue for the second time, following the 1996 tournament at the original stadium.

The tournament was well-received by fans and commentators, with the most goals per game in a European Championship since the introduction of the group stage, and only two goalless games. The refereeing style was also praised, with a conservative use of VAR and quick decisions made on the pitch. Belgium, France, and Sweden lost to lower-ranked sides.

### value_counts----->>>
#### Return a Series containing counts of unique values.

The resulting object will be in descending order so that the first element is the most frequently-occurring element. Excludes NA values by default.

````python
countrys =euro_df.Country.value_counts().index
# ['England', 'Italy', 'Spain', 'France', 'Germany', 'Portugal', 'Belgium','Denmark', 'Switzerland', 'Wales', 'Netherlands', 'Ukraine', 'Austria','Croatia', 'Czech Republic', 'SwitzerlandF', 'Sweden', 'Poland']

values=euro_df.Country.value_counts().values
#[7, 7, 5, 5, 5, 5, 4, 4, 2, 2, 2, 2, 2, 2, 2, 1, 1, 1]
  
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(values, labels = countrys)
  
# show plot
plt.show()
````

# Number of goals scored by country.

````python
country_goals=[]
for a in countrys:
    country_goals.append(sum(euro_df.Goals[euro_df.Country==a]))

plt.figure(figsize=(15,10))
sns.barplot(x=countrys, y=country_goals)
plt.xticks(rotation= 45)
plt.xlabel('countrys')
plt.ylabel('goals')
plt.title('total number of goals by countries')
````

# Number of players by position played.

````python
plt.figure(figsize=(15,10))
sns.barplot(x=euro_df['Position'].value_counts().index, y=euro_df['Position'].value_counts().values)
plt.xticks(rotation= 45)
plt.xlabel('position')
plt.ylabel('number of player')
plt.title('football player and position chart .')
````

````python
country1 = euro_df[euro_df.Country != 'No Country'].set_index('Match played').Country.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = country1, order=country1.value_counts().index[:18])
plt.show()
````
